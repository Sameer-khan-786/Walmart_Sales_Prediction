# Data Cleaning and Handling Missing Values
## Missing values in the 'CPI' and 'Unemployment' columns of the features dataset were filled with their respective medians.


```python
features_df["CPI"].fillna(features_df["CPI"].median(),inplace=True)
features_df["Unemployment"].fillna(features_df["Unemployment"].median(),inplace=True)
```

## Negative values in the 'MarkDown' columns were replaced with zeros, and missing values in these columns were also filled with zeros.

```python
for i in range(1,6):
  features_df["MarkDown"+str(i)] = features_df["MarkDown"+str(i)].apply(lambda x: 0 if x < 0 else x)
  features_df["MarkDown"+str(i)].fillna(value=0,inplace=True)
```

# Sales Trends Over Time

```python
# Aggregating the data on a monthly basis
monthly_data = df.groupby(pd.Grouper(key='Date', freq='M')).agg({'Weekly_Sales':'sum'})

# Decomposition of aggregated sales data
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(monthly_data['Weekly_Sales'], model='additive')

# Plotting the components
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(12, 8))

result.observed.plot(ax=ax1)
ax1.set_ylabel('Observed')
ax1.set_title('Seasonal Decomposition of Monthly Sales Data')

result.trend.plot(ax=ax2)
ax2.set_ylabel('Trend')

result.seasonal.plot(ax=ax3)
ax3.set_ylabel('Seasonal')

result.resid.plot(ax=ax4)
ax4.set_ylabel('Residual')

plt.tight_layout()
plt.show()
```
